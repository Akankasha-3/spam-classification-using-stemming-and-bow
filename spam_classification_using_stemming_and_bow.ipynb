# Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report,accuracy_score
```

Read data

```python
df=pd.read_csv("/content/sms_spam.csv")
df.head()
```

Import libraries that are required for spam detection

```python
import nltk
nltk.download('stopwords')
from nltk.stem import PorterStemmer
```


preprocessing and stemming

```python
for i in range(df.shape[0]):
    words=df["text"][i]
    words=re.sub('[^a-zA-Z]',' ',words)#helps in removing non alphabetics
    words=words.lower()#lowering the case
    words=words.split()
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]#stemmin and removing stopwords
    df["text"][i]=' '.join(words)
```

```python
x=df.iloc[:,1].values#independent features
y=df.iloc[:,0].values#dependent features
```

Splitting into and training and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#applying labelencoding to dependent variables
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_train=le.fit_transform(y_train)
y_test=le.transform(y_test)
```

```python
#importing and initializing logistic regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
#iporting and transforming independent data into vectors
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x_train1=cv.fit_transform(x_train)
x_test1=cv.transform(x_test)
lr.fit(x_train1,y_train)
```

Evaluating

```python
y_pred1=lr.predict(x_test1)
print(classification_report(y_pred1,y_test))
print(accuracy_score(y_pred1,y_test))
```
